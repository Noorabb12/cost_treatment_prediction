```python
# Specify the numerical column you want to analyze
numerical_column = 'charges'

# Calculate the IQR (Interquartile Range)
Q1 = df[numerical_column].quantile(0.25)
Q3 = df[numerical_column].quantile(0.75)
IQR = Q3 - Q1

# Define a threshold for outliers (e.g., 1.5 times the IQR)
outlier_threshold = 1.5 * IQR

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - outlier_threshold
upper_bound = Q3 + outlier_threshold

# Identify outliers
outliers = df[(df[numerical_column] < lower_bound) | (df[numerical_column] > upper_bound)]

# Calculate the percentage of outliers
percentage_outliers = (len(outliers) / len(df)) * 100

# Print the percentage of outliers
print(f'Percentage of outliers in "{numerical_column}": {percentage_outliers:.2f}%')
```

```python
# Create a column indicating obesity: bmi >= 30
df['bmi_obese'] = df['bmi'].apply(lambda x: 1 * (x >= 30))
```

### Data Modeling

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Holdout sample
X = df.drop(['charges'], axis=1)
y = df[['charges']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)
train_data = pd.concat([X_train, y_train], axis=1)
```

```python
model = ols("charges ~ age + smoker * bmi_obese", train_data).fit()
results = model.summary()
results
```

1. R2= 0.858. Not bad
2. The results confirm that obesity has a significant impact among smokers but not non-smokers.

```python
# Prediction performance
y_test_pred = model.predict(X_test)
model_mse = metrics.mean_squared_error(y_test, y_test_pred)
model_mae = metrics.mean_absolute_error(y_test, y_test_pred)
print(f"Mean Squared Error  = {model_mse:12.2f}")
print(f"Mean Absolute Error = {model_mae:12.2f}")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8), 
                        layout='tight', sharey=False)
sns.boxplot(x=X_train['sex'], y=model.resid, showfliers=False, ax=axs[0,0])
sns.boxplot(x=X_train['children'], y=model.resid, showfliers=False, ax=axs[0,1])
sns.regplot(x=X_train['bmi'], y=model.resid, line_kws={'color':'tab:orange'}, ax=axs[1,0])
sns.boxplot(x=X_train['region'], y=model.resid, showfliers=False, ax=axs[1,1])
plt.suptitle('Residuals')
plt.show()

```

Residuals seem to hold an approximately linear relationship with children, the number of dependents. Let's add it to the model.

```python
#### The Second Model
model2 = ols("charges ~ age + smoker * bmi_obese + children", train_data).fit()
model2.summary()
```

```python
# Prediction performance
y_test_pred = model2.predict(X_test)
model_update_mse = metrics.mean_squared_error(y_test, y_test_pred)
model_update_mae = metrics.mean_absolute_error(y_test, y_test_pred)
print(f"Mean Squared Error  = {model_update_mse:12.2f}")
print(f"Mean Absolute Error = {model_update_mae:12.2f}")
```

```python
y_test_pred = model2.predict(X_test)
ax = sns.scatterplot(x=y_test['charges'], y=y_test_pred.values)
sns.lineplot(x=y_test['charges'], y=y_test['charges'], color='tab:orange', ax=ax)
ax.set(xlabel='Actual charges', ylabel='Predicted charges',
       title='Prediction performance on the test sample')
plt.show()
```
